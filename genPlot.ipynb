```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
d_min, d_max, d_res = [-0.5, 1.5, 100]
d = np.linspace(d_min, d_max, num=d_res)
eigData = np.zeros_like(d)
rwaData = np.zeros_like(d)
qutData = np.zeros_like(d)
```

```python
for i in range(d_res):
    data = np.load(os.path.join('plotData', str(i)+'.npy'))
    eigData[i] = np.amax(data[-50:,2])
    data = np.load(os.path.join('plotDataRWA', str(i)+'.npy'))
    rwaData[i] = np.amax(data[-50:,2])
    data = np.load(os.path.join('plotDataQuTiP', str(i)+'.npy'))
    qutData[i] = np.amax(abs(data[-50:,2]))
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(d, eigData, label='eigSolve')
plt.plot(d, rwaData, label='eigSolve (RWA)')
plt.plot(d, qutData, label='QuTiP (RWA)')
plt.title(r'Drive Detuning vs Intercavity Amplitude; $\chi = 6.3\chi_1$')
plt.xlabel(r'Drive Detuning, $(\omega_d - \omega_c)/|\chi|$')
plt.ylabel(r'Intercavity Amplitude $|\langle a\rangle|$')
plt.xlim(-0.5, 1.5)
plt.legend()
plt.savefig('plot.png')
plt.show()
```
