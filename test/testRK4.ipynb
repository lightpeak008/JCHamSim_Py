```python
import numpy as np
from math import cos, sqrt
from qu_eig import *
import matplotlib.pyplot as plt
```

```python
n = 100
omega_c = (2*np.pi) * 1.0
delta   = (2*np.pi) * 0.5
omega_d = (2*np.pi) * 1.2
g       = (2*np.pi) * 0.25
xi      = (2*np.pi) * 1
kappa   = (2*np.pi) * 0.01
t_end = 10

omega_q = omega_c + delta
```

```python
I = identity(n)
a = destroy(n)
sz = -1.*I
#N = dag(a).dot(a) + sz/2 + 0.5
N = dag(a).dot(a)

D = np.sqrt((delta**2)*I + (4*g**2)*N)

H0 = omega_c*(dag(a).dot(a)) + (omega_c*I-D)*(sz/2)
Hd = (a + dag(a))

def Hd_coeff(t, args):
    xi = args['xi']
    omega_d = args['omega_d']
    return xi*np.cos(omega_d*t)/sqrt(2)
```

```python
H = [H0, [Hd, Hd_coeff]]
Hd_args = {'xi' : xi, 'omega_d' : omega_d}
H_args = [Hd_args]

psi0 = fock(n, 0)

c_op = sqrt(kappa)*a
#c_op = zeros(n)

e_op = a
```

```python
cyclePeriod = 2*np.pi/omega_d
cycleRes = 200
cycleCount = int(t_end/cyclePeriod)+1
```

```python
%time data = rk4Solve(H, psi0, cyclePeriod, cycleRes, cycleCount, c_op, e_op,\
                      H_args, ntraj=100)
np.save('testRK4_data.npy', data)
```

```python
fig = plt.figure(figsize=(10,6))
plt.title('Driven JC Oscillator: rk4Solve')
plt.xlabel('Time')
plt.ylabel(r'$\langle a^\dagger a\rangle$')
plt.plot(data[:,0], data[:,1])
plt.show()

fig = plt.figure(figsize=(10,6))
plt.title('Driven JC Oscillator: rk4Solve')
plt.xlabel('Time')
plt.ylabel(r'$\langle \|a\|\rangle$')
plt.plot(data[:,0], data[:,2])
plt.show()
```
